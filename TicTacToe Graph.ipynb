```python
import csv
import pandas as pd
df = pd.read_csv('/Users/leunghon/Desktop/tictactoe_week5/games.csv')
```

# Who win the game with the fewest steps?

```python
df.plot(x='Winner', y='Total Moves', kind='bar')
```

# Who won the game the most?

```python
df['Winner'].value_counts().plot(kind='pie')
```

# What is the average moves that needed to win the game?

```python
df['Total Moves'].plot.box(y='Winner')
```

# The average move to win for each player

```python
df.boxplot(by = 'Winner', column = ['Total Moves'], grid = False)
```
